```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re


pd.options.display.max_colwidth = 100000
```

```python
df = pd.read_csv("reviews_popular_author.csv")
stop_words = ["much", "many", "wa"] + list(STOPWORDS)
```

```python
df = df.set_index("review_id")
df = df.drop(["Unnamed: 0"], axis=1)
```

```python
fin = df.groupby(['story_id'])['content'].apply(lambda text: ''.join(text.to_string(index=False))).str.replace('(\\n)', ' ').reset_index()
fin['content'] = fin['content'].apply(lambda text: re.sub(r'[^\w\s]','',text).lower()[10:])
```

```python
def createidf(df):
    idf = {}
    for value in range(df.shape[0]):
        reviews = df.iloc[value]['content']
        for word in str(reviews).split():
            if word in idf and idf[word] <= value+1:
                    idf[word] = idf[word] + 1
            else:
                idf[word] = 1
    for value in idf:
        idf[value] = np.log(df.shape[0] / float(idf[value]))
    return(idf)

```

```python
def cloudidf(id, df, idf):
    idf = createidf(df)
    row = df.loc[df['story_id'] == id]
    wc = {}
    wl = str(row['content']).split()
    for item in wl:
        if item:
            if item in wc:  
                wc[item] += 1  
            else:  
                wc[item] = 1   
    length = len(wl)
    for k,v in list(wc.items()):
        if v < 4:
            del wc[k]
    for key in wc:
        if not key in idf:
            idf[key] = 1.0 / df.shape[0]
        wc[key] = wc[key] / idf[key]
    story = id
    wordcloud = WordCloud(max_font_size=50, max_words=50, background_color = "white").generate_from_frequencies(wc)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

    image = "./img/" + str(story) + "2.png"
    wordcloud.to_file(image)
```

```python
def cloud(id, df):
    row = df.loc[df['story_id'] == id]
    wc = {}
    wl = str(row['content']).split()
    for item in wl:
        if item:
            if item in wc:  
                wc[item] += 1  
            else:  
                wc[item] = 1
    story = id
    wordcloud = WordCloud(max_font_size=50, max_words=50, background_color = "white", stopwords = STOPWORDS).generate_from_frequencies(wc)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

    image = "./img/" + str(story) + ".png"
    wordcloud.to_file(image)
```

```python
cloud(808358, fin)
```
